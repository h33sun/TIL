### 브랜치

- branch: 가지
- git branch: 본 repo에서 나온 새로운 분기점을 의미
- 실험장이라고 봐도 무방, 실험 성공시 본 source에 merge  <br>

    - Fork: 타인의 repo를 사본으로 만들어 내 소유로 사본을 만듬
    - `git branch`: `master` 몸통을 의미, 사용자가 보게 될 가장 마지막 단계
    - `git branch (브랜치명)`: 새로운 브랜치를 만들고 이동한다.
    - `git checkout (브랜치명 or 커밋ID)`: 브랜치 이동
    - `git checkout -b (브랜치명)`: 브랜치가 없을 경우 생성 후 이동 (오타나면 짜증남)
    
    
- 상대방의 repo를 fork한다.
- fork한 나의 repo를 clone한다.

- 보통 develop 브랜치보단 develop 브랜치 내 feature 별 브랜치를 만든다.  

- 브랜치에서 더 이상 commit 할 것이 없을 경우 브랜치 이동이 가능하다.

#### - feature에서 develop으로 merge하기 (cd = develop)
    
   - `git merge (브랜치명)`: develop에서 feature commit을 올린다.
   - develop에서 추가 작업을 했을 경우, branch와 develop이 충돌이 나게 된다. msg = both merged
   
   
   
   
   - 충돌 파일을 열어보면 HEAD(현 상태) / 아래부분은 다른 버전이 나온다.
   
   
   
   
   - 충돌 부분을 수정하고 임시 캐릭터들을 지워준다.

#### 원 repo로 나의 수정 사항을 올리기?

- 1. 나의 사본에 수정 사항을 push한다.
- 2. github 나의 repo에서 pull request 버튼을 클릭 (방향에 주의 : master에 보내지 말것)

### 내 브랜치 최신화하기

- `git remote add (새로만들 remote) origin주소`
- `git pull (새로만든 remote) (내branch)`




###

- 팀장의 repo를 fork 한다.
- fork 한 나의 repo를 clone 한다.
- 팀장 repo의 주소를 pmorigin 으로 추가한다. ($ git remote add pmorigin { 주소 })
- git checkout -b develop
- git checkout -b feature/{ 작업할 기능 } 하여 해당 브랜치 위에서 작업
- git add, commit, push to feature/{ 작업할 기능 }
- git checkout develop && git merge feature/{ 작업할 기능 }
- merge 된 내용을 git push origin develop
- github에서 issue 생성
- github에서 create pull request
